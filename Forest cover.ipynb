# Target and Features

### target

```python
y = train_data.Cover_Type
```

### features

```python
features = train_data.columns[0:54]
features
```

```python
x = train_data[features]
x.head()
```

### Features Statistic

```python
x.describe()
```

# Random Forest Regressor model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state = 1)
```

```python
model.fit(x,y)
```

```python
predictions = model.predict(x)
predictions
```

```python
cover = pd.DataFrame({'Actual Cover':y, 'Predicted Cover': predictions.round(0)})
cover.head()
```

```python
cover.tail()
```

```python
cover[10:100]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(cover['Actual Cover'],cover['Predicted Cover'])
plt.show()
```

```python
plt.plot(cover['Actual Cover'],cover['Actual Cover'])
plt.scatter(cover['Actual Cover'],cover['Predicted Cover'], color='m')
plt.show()
```

## Error

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y,predictions)
```
